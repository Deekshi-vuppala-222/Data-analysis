```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(df["2022"], bins=10,  color="blue")
plt.title("Distribution of 2022 Data")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

histogram  showing the distribution of data for the year 2022 across different districts, providing insights into central tendency, dispersion, and potential skewness.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x="2001", y="2009", data=df, palette="viridis")
plt.xticks(rotation=90, ha='right')
plt.xlabel("number of teachers in2001")
plt.ylabel("Number of teachers in 2009")
plt.title("graph of 2 years")
plt.show()
```

 bar chart comparing the number of teachers in 2001 against the number in 2009 across different districts, providing a visual representation of changes over time.

```python
df.groupby("2022")["1989"].sum().plot.pie(autopct="%1.1f%%", figsize=(8, 8), cmap="Set3")
plt.title("Total")
plt.ylabel("") 
plt.show()
```

A pie chart illustrating the proportion of total '1989' values contributed by each '2022' category.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x="2020", y="1999", data=df, palette="Set2")
plt.xticks(rotation=90, ha='right')
plt.title("Number of teachers")
plt.show()
```

box plot visualizing the distribution of '1999' data for each '2020' category, illustrating the median, quartiles, and potential outliers in each group.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='2002', y='2022', data=df)
plt.title("Comparison of Teachers in 2001 vs 2022")
plt.show()
```

bar plot compares teacher counts in 2002 and 2022 across different districts, visually highlighting changes over two decades.

```python
plt.figure(figsize=(8, 6))
sns.boxplot(df["2022"])
plt.title("Box Plot for Teachers in 2022")
plt.show()
```

box plot displays the distribution of teacher counts in 2022, showing the median and potential outliers for all districts combined.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df["2001"], y=df["2022"], color="red")
plt.title("Scatter Plot of Teachers in 2001 vs 2022")
plt.xlabel("Teachers in 2001")
plt.ylabel("Teachers in 2022")
plt.show()
```

Scatter plot visualizing the relationship between teacher counts in 2001 and 2022, with each point representing a district and its teacher numbers in those years, colored in red.

```python
plt.figure(figsize=(10, 6))
plt.plot(df.columns[1:], df.iloc[5, 1:], marker="o", linestyle="-", color="green")
plt.title("Trend of Teachers in a Sample School")
plt.xlabel("Years")
plt.ylabel("Number of Teachers")
plt.show()
```

A line graph depicting the trend in teacher numbers over the years for Jalandhar (row 5), illustrated with a green line and markers to clearly indicate teacher counts for each year.

```python
top_5 = df.sort_values("2022", ascending=False).head(5)
plt.figure(figsize=(8, 8))
plt.pie(top_5["2022"], labels=top_5.iloc[:, 0], autopct="%1.1f%%", startangle=140)
plt.title("Top 5 Schools with Most Teachers in 2022")
plt.show()
```

A pie chart showing the distribution of the top 5 districts with the highest number of teachers in 2022, with each slice representing a district's proportion of the total teacher count among those top 5 districts.

```python
plt.figure(figsize=(8, 6))
sns.countplot(x=pd.cut(df["2022"], bins=5))
plt.title("Categorized Count of Teachers in 2022")
plt.xlabel("Teacher Count Ranges")
plt.show()
```

A count plot showing the number of districts falling into each of the five bins (ranges) of teacher counts in 2022, providing a categorized view of teacher distribution across districts.

area plot showing the trend of the average number of teachers across all districts over the years, with the shaded area representing the mean teacher count and a light blue fill to visualize the overall temporal change.
